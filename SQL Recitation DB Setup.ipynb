```python
import pandas as pd
import numpy as np
import sqlite3
db = sqlite3.connect('Recitation')
c= db.cursor()
```

```python
stmt= ('CREATE TABLE Student ' +
       '(sid INTEGER NOT NULL, ' +
        'name CHAR(20), ' +
        'PRIMARY KEY (sid)); ')
c.execute(stmt)
```

```python
c.execute('INSERT INTO Student (sid,name) VALUES (1, \"Jill\");')
c.execute('INSERT INTO Student (sid,name) VALUES (2, \"Bo\");')
c.execute('INSERT INTO Student (sid,name) VALUES (3, \"Maya\");')
```

```python
stmt= ('CREATE TABLE Course ' +
       '(cid INTEGER NOT NULL, ' +
        'sem CHAR(5), ' +
        'subj CHAR(20),' +
        'PRIMARY KEY (cid, sem)); ')
c.execute(stmt)
```

```python
c.execute('INSERT INTO Course (cid,sem,subj) VALUES (550, \"S15\", \"DB\");')
c.execute('INSERT INTO Course (cid,sem,subj) VALUES (677, \"S15\", \"Algo\");')
c.execute('INSERT INTO Course (cid,sem,subj) VALUES (521, \"F15\", \"AI\");')
c.execute('INSERT INTO Course (cid,sem,subj) VALUES (700, \"S17\", \"BigData\");')
```

```python
stmt= ('CREATE TABLE Professor ' +
       '(fid INTEGER NOT NULL, ' +
       ' name CHAR(20), ' +
        'PRIMARY KEY (fid)); ')
c.execute(stmt);
```

```python
c.execute('INSERT INTO Professor (fid,name) VALUES (11, \"Ives\");')
c.execute('INSERT INTO Professor (fid,name) VALUES (12, \"Khanna\");')
c.execute('INSERT INTO Professor (fid,name) VALUES (15, \"Davidson\");')
c.execute('INSERT INTO Professor (fid,name) VALUES (18, \"Marcus\");')
```

```python
stmt= ('CREATE TABLE Teaches ' +
       '(fid INTEGER NOT NULL, ' +
       ' cid INTEGER NOT NULL, ' +
        'sem CHAR(5), ' +
        'PRIMARY KEY (cid, sem)); ')
c.execute(stmt)
```

```python
c.execute('INSERT INTO Teaches (fid,cid,sem) VALUES (11, 550, \"S15\");')
c.execute('INSERT INTO Teaches (fid,cid,sem) VALUES (12, 677, \"S15\");')
c.execute('INSERT INTO Teaches (fid,cid,sem) VALUES (18, 521, \"F15\");')
c.execute('INSERT INTO Teaches (fid,cid,sem) VALUES (11, 700, \"S17\");')
c.execute('INSERT INTO Teaches (fid,cid,sem) VALUES (15, 550, \"F16\");')
```

```python
stmt= ('CREATE TABLE Takes ' +
       '(sid INTEGER NOT NULL, ' +
       ' cid INTEGER NOT NULL, ' +
        'sem CHAR(5), ' +
        'grade CHAR(2),' +
        'PRIMARY KEY (sid, cid, sem)); ')
c.execute(stmt)
```

```python
c.execute('INSERT INTO Takes (sid,cid,sem,grade) VALUES (1, 550, \"S15\", \"A\");')
c.execute('INSERT INTO Takes (sid,cid,sem,grade) VALUES (1, 677, \"S15\", \"A\");')
c.execute('INSERT INTO Takes (sid,cid,sem,grade) VALUES (3, 550, \"S15\", \"C\");')
c.execute('INSERT INTO Takes (sid,cid,sem,grade) VALUES (3, 521, \"F15\", \"C\");')
c.execute('INSERT INTO Takes (sid,cid,sem,grade) VALUES (3, 700, \"S17\", \"A\");')
```
