# Decision Trees

We will use the cars dataset from https://archive.ics.uci.edu/ml/datasets/Car+Evaluation

Also we use some code from http://www.shogun-toolbox.org/static/notebook/current/DecisionTrees.html

```python
# class attribute
evaluation = {'unacc' : 1, 'acc' : 2, 'good' : 3, 'vgood' : 4}

# non-class attributes
buying = {'vhigh' : 1, 'high' : 2, 'med' : 3, 'low' : 4}
maint = {'vhigh' : 1, 'high' : 2, 'med' : 3, 'low' : 4}
doors = {'2' : 1, '3' : 2, '4' : 3, '5more' : 4}
persons = {'2' : 1, '4' : 2, 'more' : 3}
lug_boot = {'small' : 1, 'med' : 2, 'big' : 3}
safety = {'low' : 1, 'med' : 2, 'high' : 3}
```

```python
f = open('car.data', 'r')

X = []
y = []

# read data from file and encode
for line in f:
    words = line.rstrip().split(',')
    words[0] = int(buying[words[0]])
    words[1] = int(maint[words[1]])
    words[2] = int(doors[words[2]])
    words[3] = int(persons[words[3]])
    words[4] = int(lug_boot[words[4]])
    words[5] = int(safety[words[5]])
    words[6] = evaluation[words[6]]
    X.append(words[0:6])
    y.append(int(words[6]))

f.close()
```

```python
import numpy as np
from numpy import random, delete

features = np.array(X)
labels = np.array(y)

# number of test vectors
num_test_vectors = 200;

test_indices = random.randint(features.shape[0], size = num_test_vectors)
test_features = features[test_indices]
test_labels = labels[test_indices]

# remove test vectors from training set
features = delete(features,test_indices,0)
labels = delete(labels,test_indices,0)
```

## Detailed breakdown of computing entropy, info gain


```python
# From http://gabrielelanaro.github.io/blog/2016/03/03/decision-trees.html

import numpy as np
x1 = [0, 1, 1, 2, 2, 2]
x2 = [0, 0, 1, 1, 1, 0]
y = np.array([0, 0, 0, 1, 1, 0])

def partition(a):
    return {c: (a==c).nonzero()[0] for c in np.unique(a)}

def entropy(s):
    res = 0
    val, counts = np.unique(s, return_counts=True)
    freqs = counts.astype('float')/len(s)
    for p in freqs:
        if p != 0.0:
            res -= p * np.log2(p)
    return res

def mutual_information(y, x):

    res = entropy(y)
    print ('Initial entropy: ', res)

    # We partition x, according to attribute values x_i
    val, counts = np.unique(x, return_counts=True)
    freqs = counts.astype('float')/len(x)

    # We calculate a weighted average of the entropy
    for p, v in zip(freqs, val):
        print ('Split on ', v, entropy(y[x == v]))
        res -= p * entropy(y[x == v])

    return res

from pprint import pprint

def is_pure(s):
    return len(set(s)) == 1

def recursive_split(x, y):
    # If there could be no split, just return the original set
    if is_pure(y) or len(y) == 0:
        return y

    # We get attribute that gives the highest mutual information
    gain = np.array([mutual_information(y, x_attr) for x_attr in x.T])
    selected_attr = np.argmax(gain)

    # If there's no gain at all, nothing has to be done, just return the original set
    if np.all(gain < 1e-6):
        return y


    # We split using the selected attribute
    sets = partition(x[:, selected_attr])

    res = {}
    for k, v in sets.items():
        y_subset = y.take(v, axis=0)
        x_subset = x.take(v, axis=0)

        res["x_%d = %d" % (selected_attr, k)] = recursive_split(x_subset, y_subset)

    return res

X = np.array([x1, x2]).T
pprint(recursive_split(X, y))
```
